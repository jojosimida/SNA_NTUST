```python
import pandas as pd
import numpy as np
import pickle
import codecs
```

```python
url = 'https://zh.wikipedia.org/wiki/%E4%B8%89%E5%9B%BD%E6%BC%94%E4%B9%89%E8%A7%92%E8%89%B2%E5%88%97%E8%A1%A8'
```

```python
all_name = []
all_2ndname = {}
path = '/dataset/SNA_final/'
```

```python
def get_1stname(df):
    names = df['姓名'].values
    
    # clean the name column
    ignore1 = ' ('
    ignore2 = '（'
    for i, name in enumerate(names):
        if name.find(ignore1)!=-1:
            names[i] = name[:name.find(ignore1)]
            
        elif name.find(ignore2)!=-1:
            names[i] = name[:name.find(ignore2)]
              
    return names
```

```python
def get_2ndname(df):
    dict_2ndname = {}
    second_name = df['字'].values
    firstname = get_1stname(df)
    
    # clean the second name
    ignore = '演義字'
    for i,_2ndname in enumerate(second_name):
        if pd.isnull(_2ndname)==False:
            if _2ndname.find(ignore)!=-1:
                dict_2ndname[firstname[i]] = _2ndname[_2ndname.find(ignore)+3:]

            else:
                dict_2ndname[firstname[i]] = _2ndname

    return dict_2ndname
```

```python
table = pd.read_html(url, header=0)
```

```python
for i in range(len(table)):
    name  = get_1stname(table[i])
    _2ndname = get_2ndname(table[i])
    
    all_name.extend(name)
    all_2ndname.update(_2ndname)
```
